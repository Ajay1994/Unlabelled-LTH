```python
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
def get_mean_and_std(dataset):
    '''Compute the mean and std value of dataset.'''
    dataloader = trainloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=True, num_workers=2)

    mean = torch.zeros(3).cuda()
    std = torch.zeros(3).cuda()
    print('==> Computing mean and std..')
    for batch_idx, (image_name, inputs, targets) in enumerate(trainloader):
        inputs = inputs.cuda()
#         print(image_name)
        for i in range(3):
            mean[i] += inputs[:,i,:,:].mean()
            std[i] += inputs[:,i,:,:].std()
    mean.div_(len(dataset))
    std.div_(len(dataset))
    return mean, std
```

```python
CLASS_NAMES = [
            'Atelectasis',
            'Cardiomegaly',
            'Effusion',
            'Infiltration',
            'Mass',
            'Nodule',
            'Pneumonia',
            'Pneumothorax',
            'Consolidation',
            'Edema',
            'Emphysema',
            'Fibrosis',
            'Pleural_Thickening',
            'Hernia']
```

```python
import glob
import shutil
for c in CLASS_NAMES:
    if c in df.columns:
        df_class = df[df[c] == 1.0][:10000]
        subject_id = list(df_class["subject_id"])
        study_id = list(df_class["study_id"])
        for i in range(0, len(subject_id)):
            folder = data_loc + "/p" + str(subject_id[i])[:2] + "/p" + str(subject_id[i]) + "/s" + str(study_id[i])
            for file in glob.iglob(folder + '/*.jpg', recursive=True):
                shutil.copy(file, "/data2/ajay_data/MIMIC_CXR_unlabelled/")
```
